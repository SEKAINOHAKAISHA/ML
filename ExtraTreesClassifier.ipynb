```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(100,n_features=11,random_state=21)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=16,test_size=0.2)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
etc = ExtraTreesClassifier(random_state=21)
```

```python
etc.fit(X_train,y_train)
```

```python
y_pred = etc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(etc,X_train,y_train,scoring='accuracy',cv=5,n_jobs=-1).mean()
```

```python
param_grid = {
    'n_estimators' : [100,250,500],
    'min_samples_leaf' : [5,15,25],
    'max_features' : [3,5,7,9],
    'criterion' : ['gini','entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
etc2 = GridSearchCV(etc,param_grid,cv=3,n_jobs=-1)
```

```python
etc2.fit(X_train,y_train)
```

```python
print(etc2.best_params_)
```

```python
print(etc2.best_score_)
```
