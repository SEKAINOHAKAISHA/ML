```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2022mlbteams.csv')
```

```python
df
```

```python
df.columns
```

```python
df.drop(30,inplace=True)
```

```python
df.drop(['Tm','#Bat'],axis=1,inplace=True)
```

```python
df
```

```python
X = df.iloc[:,0:27]
```

```python
y = df.iloc[:,27]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=21,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
df.columns
```

```python
X_train = pd.DataFrame(X_train,columns = ['BatAge', 'R/G', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI',
       'SB', 'CS', 'BB', 'SO', 'BA', 'OBP', 'SLG', 'OPS', 'OPS+', 'TB', 'GDP',
       'HBP', 'SH', 'SF', 'IBB', 'LOB'])
```

```python
X_train.head(10)
```

```python
X_train.describe().round(3)
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA()
```

```python
X_pca1 = pca1.fit_transform(X_train)
```

```python
pca1.explained_variance_ratio_
```

```python
#Prasad Ostwal Github, copied from there
```

```python
import matplotlib.pyplot as plt
```

```python
pca1.explained_variance_
```

```python
plt.bar(range(1,len(pca1.explained_variance_)+1),pca1.explained_variance_)
plt.xlabel('Components')
plt.ylabel('Explained variance')
plt.plot(range(1,len(pca1.explained_variance_)+1),np.cumsum(pca1.explained_variance_),
                c = 'red',
                label = 'Cumulative Explained Variance')
plt.legend(loc='upper left')
```

```python
plt.plot(pca1.explained_variance_)
plt.xlabel('Components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca1.components_
```

```python
pca2 = PCA(0.95)
```

```python
X_pca2 = pca2.fit_transform(X_train)
```

```python
X_pca2.shape
```

```python
pca2.explained_variance_ratio_
```

```python
pca2c = PCA(n_components=2)
```

```python
X_pca2c = pca2c.fit_transform(X_train)
```

```python
colormap = plt.cm.get_cmap('coolwarm')
```

```python
plt.figure()
plt.scatter(X_pca2c[:,0],X_pca2c[:,1],c = y_train,cmap = colormap)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Playoff')
plt.show()
```

```python
pca3c = PCA(n_components=3)
```

```python
X_pca3c = pca3c.fit_transform(X_train)
```

```python
# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the first 3 principal components
scatter = ax.scatter(X_pca3c[:,0], X_pca3c[:,1], X_pca3c[:,2], c=y_train, marker='o')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA of the Dataset')

plt.show()
```
