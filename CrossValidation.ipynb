```python
import pandas as pd
```

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
fastball_speed = np.random.randint(90,106,size=500)
```

```python
tommy_john = np.where(fastball_speed > 96, np.random.choice([0,1],size=500, p =[0.3,0.7]),0)
#np.where is like (condition, if conditon True then execute this part,if condition False execute this part,
# if (fastball_speed > 96):
#   then np.random.choice([0,1],size=500, p =[0.3,0.7])
# else:
#    0
```

```python
d = {'fastball_speed' : fastball_speed,
    'tommy_john' : tommy_john}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
X = df.iloc[:,0:1]
```

```python
y = df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train.values.ravel())
#ravel returns a flattened(1D) data of the same type
```

```python
lr.score(X_test,y_test)
```

```python
#test it again with different random_state
```

```python
X2_train,X2_test,y2_train,y2_test=train_test_split(X,y,random_state=25,test_size=0.2)
```

```python
lr.fit(X2_train,y2_train)
```

```python
lr.score(X2_test,y2_test)
```

```python
#basic cross validation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvscore = cross_val_score(lr,X,y.values.ravel(),cv=10)
```

```python
cvscore
```

```python
np.average(cvscore)
```

```python
np.std(cvscore)
```

```python
#define standard k fold
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=15 , shuffle=True , random_state=42)
```

```python
# cv can be Int, if it is None then StratifiedKFold is used directly in specific cases as per documentation
# or we can also assign a Cross-Validation method like in kfscore3 to cv param
```

```python
kfscore = cross_val_score(lr,X,y.values.ravel(),cv=kf,scoring='f1')
```

```python
kfscore2 = cross_val_score(lr,X,y.values.ravel(),cv=kf,scoring='accuracy')
```

```python
kfscore
```

```python
np.average(kfscore)
```

```python
np.std(kfscore)
```

```python
kfscore2
```

```python
np.average(kfscore2)
```

```python
np.std(kfscore2)
```

```python
#stratified K-Fold
#which basically has equal percentage of all classes in the training and testing data in all subsets
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=10 ,shuffle=True, random_state=11)
```

```python
kfscore3 = cross_val_score(lr,X,y.values.ravel(),cv = skf,n_jobs=-1)
```

```python
kfscore3
```

```python
np.average(kfscore3)
```

```python
np.std(kfscore3)
```
