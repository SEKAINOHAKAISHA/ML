```python
import pandas as pd
```

```python
data = pd.read_csv('500hits.csv',encoding='latin-1')
```

```python
df = data.drop(columns=['PLAYER','CS'])
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ode = OrdinalEncoder()
```

```python
df = ode.fit_transform(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame(df)
```

```python
X = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=100,solver='newton-cholesky')
#default solver lbfgs failed to converge and b,f calculation errors
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
