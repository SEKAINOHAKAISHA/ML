```python
import pandas as pd
```

```python
df = pd.read_csv('500hits.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.drop(columns=['PLAYER','CS'],inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
#Decision Trees do not need data to be Scaled or Normalized
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=17,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.get_params()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
cr = classification_report(y_test,y_pred)
```

```python
print(cr)
```

```python
dtc.feature_importances_
```

```python
features = pd.DataFrame(dtc.feature_importances_,index=X.columns)
```

```python
features
```

```python
#ccp basically carries out pruning by removing the braches of low aplha value
dtc2 = DecisionTreeClassifier(criterion='entropy',ccp_alpha=0.04)
```

```python
dtc2.fit(X_train,y_train)
```

```python
y_pred2 = dtc2.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
```

```python
print(classification_report(y_test,y_pred2))
```
